## Data scalation [0-1]

```python
scalers = {}
features = {}
for feature in df.columns[1:]:
    scaler = MinMaxScaler()
    y_scal = np.reshape(y.values,(len(y),1))
    y_scal = scaler.fit_transform(y_scal)
    scalers[feature] = scaler
    y_scal = np.reshape(y_scal,len(y))
    features[feature] = y_scal
```

## MODEL DEFINITION

```python
c = Input(shape=(WINDOW_SIZE,1),name='Close')

inputs={'Close': c}

DROPOUT_RATE = 0.2

x = LSTM(WINDOW_SIZE, kernel_initializer='random_normal',return_sequences=True)(c)
x = Dropout(DROPOUT_RATE)(x)
x = LSTM(WINDOW_SIZE, kernel_initializer='random_normal',return_sequences=True)(x)
x = Dropout(DROPOUT_RATE)(x)
x = LSTM(WINDOW_SIZE, kernel_initializer='random_normal',return_sequences=True)(x)
x = Dropout(DROPOUT_RATE)(x)
#x = Concatenate(axis=1)([x1,x2,x3,x4,x5])

x = Dense(WINDOW_SIZE, kernel_initializer='random_normal',activation='relu')(x)
# x = Dense(WINDOW_SIZE, kernel_initializer='random_normal',activation='relu')(x)
# x = Dense(WINDOW_SIZE, kernel_initializer='random_normal',activation='relu')(x)

x = Flatten()(x)
# x = Dense(7, kernel_initializer='random_normal',activation='relu')(x)
# x = Dense(7, kernel_initializer='random_normal',activation='relu')(x)

output = Dense(1, kernel_initializer='random_normal',activation='relu')(x)

model = Model(inputs=inputs,outputs=output)

model.compile(optimizer=adam, loss='mse')
```

### Get training set

```python
X_train = {'Close':[],
           'Open':[],
           'Low':[],
           'High':[],
           'Adj Close':[]}
Y_train = {'Close':[]}

TRAIN_RANGE_MIN = WINDOW_SIZE
TRAIN_RANGE_MAX = round(TRAIN_PTG*len(df))

print('Training set range: {} to {}'.format(TRAIN_RANGE_MIN,TRAIN_RANGE_MAX-1))

for i in range(TRAIN_RANGE_MIN,TRAIN_RANGE_MAX):
    
    if len(X_train['Close'])==0:
        X_train['Close'] = np.array([features['Close'][i-WINDOW_SIZE:i]])
        X_train['Open'] = np.array([features['Open'][i-WINDOW_SIZE:i]])
        X_train['Low'] = np.array([features['Low'][i-WINDOW_SIZE:i]])
        X_train['High'] = np.array([features['High'][i-WINDOW_SIZE:i]])
        X_train['Adj Close'] = np.array([features['Adj Close'][i-WINDOW_SIZE:i]])
        
    else:
        for key in X_train.keys():
            window = features[key][i-WINDOW_SIZE:i]
            X_train[key] = np.append(X_train[key], np.reshape(window,(1,WINDOW_SIZE)),axis=0)
        
    Y_train['Close'] = np.append(Y_train['Close'],features['Close'][i])
```

### Get validation set

```python
X_val = {'Close':[],
           'Open':[],
           'Low':[],
           'High':[],
           'Adj Close':[]}
Y_val = {'Close':[]}

VAL_RANGE_MIN = round(TRAIN_PTG*len(df))
VAL_RANGE_MAX = round((VAL_PTG+TRAIN_PTG)*len(df))

print('Validation set range: {} to {}'.format(VAL_RANGE_MIN,VAL_RANGE_MAX-1))

for i in range(VAL_RANGE_MIN,VAL_RANGE_MAX):
     
    if len(X_val['Close'])==0:
        X_val['Close'] = np.array([features['Close'][i-WINDOW_SIZE:i]])
        X_val['Open'] = np.array([features['Open'][i-WINDOW_SIZE:i]])
        X_val['Low'] = np.array([features['Low'][i-WINDOW_SIZE:i]])
        X_val['High'] = np.array([features['High'][i-WINDOW_SIZE:i]])
        X_val['Adj Close'] = np.array([features['Adj Close'][i-WINDOW_SIZE:i]])
        
    else:
        for key in X_train.keys():
            window = features[key][i-WINDOW_SIZE:i]
            X_val[key] = np.append(X_val[key], np.reshape(window,(1,WINDOW_SIZE)),axis=0)
        
    Y_val['Close'] = np.append(Y_val['Close'],features['Close'][i])
```

### Get test set

```python
X_test = {'Close':[],
           'Open':[],
           'Low':[],
           'High':[],
           'Adj Close':[]}
Y_test = {'Close':[]}

TEST_RANGE_MIN = round((VAL_PTG+TRAIN_PTG)*len(df))
TEST_RANGE_MAX = len(df)

print('Test set range: {} to {}'.format(TEST_RANGE_MIN,TEST_RANGE_MAX-1))

for i in range(TEST_RANGE_MIN,TEST_RANGE_MAX):
     
    if len(X_test['Close'])==0:
        X_test['Close'] = np.array([features['Close'][i-WINDOW_SIZE:i]])
        X_test['Open'] = np.array([features['Open'][i-WINDOW_SIZE:i]])
        X_test['Low'] = np.array([features['Low'][i-WINDOW_SIZE:i]])
        X_test['High'] = np.array([features['High'][i-WINDOW_SIZE:i]])
        X_test['Adj Close'] = np.array([features['Adj Close'][i-WINDOW_SIZE:i]])
        
    else:
        for key in X_train.keys():
            window = features[key][i-WINDOW_SIZE:i]
            
            X_test[key] = np.append(X_test[key], np.reshape(window,(1,WINDOW_SIZE)),axis=0)
        
    Y_test['Close'] = np.append(Y_test['Close'],features['Close'][i])
```
